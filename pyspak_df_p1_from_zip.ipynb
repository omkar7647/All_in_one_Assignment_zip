**Q36.** Assign a row number to each timeseries data record within `workspace_id = 53`, ordered by `dttime`. Output columns: `ts_data_id`, `ts_id`, `dttime`, `data`, `version_id`, `row_num`.

```python
from pyspark.sql.window import Window

window_36= Window.orderBy("dttime")

in_timeseries_data_df.join(in_timeseries_version_df,on="version_id",how="inner") \
        .select("ts_data_id", "ts_id", "dttime", "data", "version_id") \
            .withColumn("row_num", F.row_number().over(window_36)).show()
```

**Q37.** Rank timeseries data values within each `version_id` for `workspace_id = 53`, ordered by `data` value descending. Output columns: `ts_data_id`, `ts_id`, `data`, `version_id`, `rank`.

```python
window_37=Window.orderBy(F.col("data").desc())

in_timeseries_data_df.withColumn("rank",F.rank().over(window_37)) \
    .select("ts_data_id","ts_id","data","version_id","rank").show()
```

**Q38.** Calculate the running total (cumulative sum) of `data` values for each `ts_id`, ordered by `dttime`, for `workspace_id = 53`. Output columns: `ts_data_id`, `ts_id`, `dttime`, `data`, `version_id`, `running_total`.

```python
window_38=Window.orderBy(F.col("dttime")).rowsBetween(Window.unboundedPreceding, Window.currentRow)

in_timeseries_data_df.withColumn("running_total",F.sum("data").over(window_38)).show()
```

**Q39.** Find the previous and next `data` values for each timeseries record ordered by `dttime`, where `workspace_id = 53` and `version_id = 125`. Output columns: `ts_data_id`, `ts_id`, `dttime`, `data`, `previous_value`, `next_value`.

```python
window_39=Window.orderBy("dttime")

in_timeseries_data_df.filter(F.col("version_id")==125).withColumn("next_data",F.lead("data").over(window_39)) \
    .withColumn("prev_data",F.lag("data").over(window_39)).show()
```

**Q40.** Calculate the moving average of `data` values over the last 3 records (including current) for each timeseries ordered by `dttime`, where `workspace_id = 53`. Output columns: `ts_data_id`, `ts_id`, `dttime`, `data`, `moving_avg_3`.

```python

window_40 = Window.orderBy("dttime").rowsBetween(-2, 0)

in_timeseries_data_df.withColumn("avg_over_last_3", F.avg("data").over(window_40)).show()
```

**Q41.** Assign a dense rank to materials based on the count of their timeseries records in `workspace_id = 53`. Output columns: `material_id`, `timeseries_count`, `dense_rank`.

```python

material_count_df =in_timeseries_data_df.join(in_timeseries_df, on="ts_id", how="inner")\
                            .join(in_resource_df.filter(F.col("workspace_id") == 53),on="geography_id",how="inner")

material_df = material_count_df.groupBy("material_id").agg(F.count("*").alias("timeseries_count"))

window_41 = Window.orderBy(F.col("timeseries_count").desc())

material_df.withColumn("dense_rank", F.dense_rank().over(window_41)).show()
```

**Q42.** Partition timeseries data by `version_id` and display the maximum `data` value within each version for `workspace_id = 53`. Output columns: `ts_data_id`, `ts_id`, `dttime`, `data`, `version_id`, `max_in_version`.

```python

in_timeseries_data_df.groupBy("version_id").agg(F.max("data").alias("max_in_version")).show()

```

**Q43.** Calculate the percentage contribution of each `data` value to the total sum of its version for `workspace_id = 53`. Output columns: `ts_data_id`, `ts_id`, `data`, `version_id`, `percentage_of_total`.

```python
window_spec = Window.partitionBy("version_id")


in_timeseries_data_df.withColumn("version_total", F.sum("data").over(window_spec)) \
    .withColumn("percentage_of_total",(F.col("data") / F.col("version_total")) * 100) \
        .select("ts_data_id","ts_id","data","version_id","percentage_of_total").show()
```

**Q44.** Find the first and last `data` values for each `ts_id` ordered by `dttime`, where `workspace_id = 53`. Output columns: `ts_data_id`, `ts_id`, `dttime`, `data`, `first_value`, `last_value`.

```python

df=in_timeseries_data_df.join(in_timeseries_df, on="ts_id", how="inner") \
    .join(in_resource_df.filter(F.col("workspace_id") == 53),on="geography_id",how="inner")

window_spec =Window.partitionBy("ts_id").orderBy("dttime").rowsBetween(Window.unboundedPreceding, Window.unboundedFollowing)

df.withColumn("first_value", F.first("data").over(window_spec)).withColumn("last_value", F.last("data").over(window_spec)) \
    .select("ts_data_id","ts_id","dttime","data","first_value","last_value").show()
```

**Q45.** Calculate the difference between each `data` value and the average `data` value within its version for `workspace_id = 53`. Output columns: `ts_data_id`, `ts_id`, `data`, `version_id`, `avg_in_version`, `difference_from_avg`.

```python

df =in_timeseries_data_df.join(in_timeseries_df, on="ts_id", how="inner") \
    .join(in_resource_df.filter(F.col("workspace_id") == 53),on="geography_id",how="inner")

window_spec = Window.partitionBy("version_id")

df.withColumn("avg_in_version", F.avg("data").over(window_spec)) \
    .withColumn("difference_from_avg",F.col("data") - F.col("avg_in_version"))\
        .select("ts_data_id","ts_id","data","version_id","avg_in_version","difference_from_avg").show()
```

**Q46.** Rank materials by their total `data` sum across all timeseries in `workspace_id = 53`. Output columns: `material_id`, `material_name`, `total_data_sum`, `rank`.

```python

df = in_timeseries_data_df.join(in_timeseries_df, on="ts_id", how="inner")  \
    .join(in_resource_df.filter(F.col("workspace_id") == 53),on="geography_id",how="inner") \
    .join(in_material_df, on="material_id", how="inner")

material_sum_df =df.groupBy("material_id", "material_name").agg(F.sum("data").alias("total_data_sum"))

rank_window = Window.orderBy(F.col("total_data_sum").desc())

material_sum_df.withColumn("rank", F.dense_rank().over(rank_window)).orderBy("rank").show()

```

**Q47.** Display all timeseries data records where the `data` value is greater than the average `data` value across all records in `workspace_id = 53`. Output columns: `ts_data_id`, `ts_id`, `dttime`, `data`, `version_id`.

```python

avg_data = in_timeseries_data_df.agg(F.avg("data").alias("avg_data")).collect()[0]["avg_data"]

df.filter(F.col("data") > avg_data).select("ts_data_id","ts_id","dttime","data","version_id").show()
```

**Q48.** Find the material with the maximum number of timeseries records in `workspace_id = 53`. Output columns: `material_id`, `material_name`, `timeseries_count`.

```python

df = in_timeseries_data_df.join(in_timeseries_df, on="ts_id", how="inner") \
    .join(in_resource_df.filter(F.col("workspace_id") == 53),on="geography_id",how="inner") \
    .join(in_material_df, on="material_id", how="inner")

material_counts_df = df.groupBy("material_id", "material_name").count().withColumnRenamed("count", "timeseries_count")

material_counts_df.orderBy( F.col("timeseries_count").desc()).limit(1).show()
```
