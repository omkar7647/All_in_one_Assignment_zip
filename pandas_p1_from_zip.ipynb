**Q38.** Calculate the running total (cumulative sum) of `data` values for each `ts_id`, ordered by `dttime`, for `workspace_id = 53`. Output columns: `ts_data_id`, `ts_id`, `dttime`, `data`, `version_id`, `running_total`.

â†’ in_resource        (geography_id â†’ workspace_id)


```python
result = in_timeseries_data.merge(in_timeseries,on="ts_id", how="inner") \
           .merge(in_resource[in_resource["workspace_id"] == 53],on="geography_id",how="inner") \
           .sort_values(["ts_id", "dttime"]) 

result["running_total"] = result.groupby("ts_id")["data"].cumsum()

result[["ts_data_id","ts_id","dttime","data","version_id","running_total"]]
```

**Q39.** Find the previous and next `data` values for each timeseries record ordered by `dttime`, where `workspace_id = 53` and `version_id = 125`. Output columns: `ts_data_id`, `ts_id`, `dttime`, `data`, `previous_value`, `next_value`.

in_timeseries_data
â†’ in_timeseries      (ts_id â†’ version_id, geography_id)
â†’ in_resource        (geography_id â†’ workspace_id)


```python
result =in_timeseries_data.merge(in_timeseries,on="ts_id",how="inner")  \
           .merge(in_resource[in_resource["workspace_id"] == 53],on="geography_id",how="inner")

result = result[result["version_id"] == 125].sort_values(["ts_id", "dttime"])

result["previous_value"] = result.groupby("ts_id")["data"].shift(1)
result["next_value"] = result.groupby("ts_id")["data"].shift(-1)

result.sort_index()[["ts_data_id","ts_id","dttime","data","previous_value","next_value"]]


```

**Q40.** Calculate the moving average of `data` values over the last 3 records (including current) for each timeseries ordered by `dttime`, where `workspace_id = 53`. Output columns: `ts_data_id`, `ts_id`, `dttime`, `data`, `moving_avg_3`.

in_timeseries_data
â†’ in_timeseries      (ts_id â†’ geography_id)
â†’ in_resource        (geography_id â†’ workspace_id)


```python
result =in_timeseries_data.merge(in_timeseries,on="ts_id",how="inner")  \
            .merge(in_resource[in_resource["workspace_id"] == 53],on="geography_id",how="inner")  \
            .sort_values(["ts_id", "dttime"])

result["moving_avg_3"] = result.groupby("ts_id")["data"].rolling(window=3, min_periods=1).mean().reset_index(level=0, drop=True)

result.sort_index()[["ts_data_id", "ts_id", "dttime", "data", "moving_avg_3"]]
```
